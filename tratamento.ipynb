```python
columns_to_drop = ["Related_Mange", "staff", "Japanese_name", "Content_Warning", "End_year"]
df_treat= df_anime.drop(columns=columns_to_drop)
#filtered_df = df_anime[df_anime['Japanese_name'] == 'Love Live Superstar!!']
df_treat.head(3)
```

```python
df_anime[df_anime['Tags'].isnull()]
df_treated = df_treat.dropna(subset=['Tags'])
df_treated = df_treated.dropna(subset=['Rating'])
df_treated = df_treated.dropna(subset=['Studio'])
df_treated = df_treated.dropna(subset=['Voice_actors'])

df_treated_rating = df_treated.drop(df_treated[df_treated['Rating'] < 3.01].index)

df_treated_rating
```

```python
#FUNCIONA
#Função para remover os elementos que não são voice actors
def split_string(string):
    last_colon_index = string.rfind(':')
    first_comma_index = string.find(',', last_colon_index)
    
    if first_comma_index != -1:
        split_string = string[:first_comma_index].strip()
        return split_string
    
    return string
```

```python
df_seiyuu = df_treated_rating[['Rank', 'Name', 'Voice_actors']]

df_seiyuu['pairs'] = df_seiyuu['Voice_actors'].apply(lambda x: x.split(','))

# Explode the 'pairs' column to create multiple rows for each value
df_seiyuu = df_seiyuu.explode('pairs').reset_index(drop=True)

# Display the resulting dataframe
df_seiyuu[['char', 'seiyuu']] = df_seiyuu['pairs'].str.split(':', n=1, expand=True)

df_seiyuu = df_seiyuu[['Name', 'char' , 'seiyuu']]
```

```python
df_exploded = df_treated_rating.copy()

df_exploded['pairs'] = df_exploded['Voice_actors'].apply(lambda x: x.split(','))

# Explode the 'pairs' column to create multiple rows for each value
df_exploded = df_exploded.explode('pairs').reset_index(drop=True)

# Display the resulting dataframe
df_exploded[['char', 'seiyuu']] = df_exploded['pairs'].str.split(':', n=1, expand=True)

df_exploded
```

```python
df_exploded= df_exploded.dropna(subset=['seiyuu'])

df_exploded
```
